### Average performance of GBTs


```python
# Make sure to execute this only ones
metrics = ["train_r2","ev20_r2","ev100_r2","evTex_r2",
           "train_f2","ev20_f2" ,"ev100_f2" ,"evTex_f2",
           "train_recall","ev20_recall" ,"ev100_recall","evTex_recall",
           'ev_elmod_r2', 'ev_osf_france_r2', 'ev_osf_gb_r2','ev_osf_spain_r2', 'ev_texas_r2'] #<- new grids
indices = [ti for ti in set(mi["train_idx"]) if not pd.isnull(ti)]
mnames = ['GradientBoostingClassifierSNBS', 'GradientBoostingClassifierTroublemaker',
          'GradientBoostingRegressionSNBS', 'GradientBoostingRegressionTroublemaker']

# Since GB performance is not really affected by more data, take global mean
for mname in mnames:
    if mname + "glomean" in set(mi["model_name"]): 
        break
    for npg in [20, 100]:
        mods = mi[(mi["model_name"] == mname) & (mi["nodes_per_grid"]==npg)]
        bestmods = mods[metrics].sort_values("ev"+str(int(npg))+"_r2").tail(3)
        task = set(mods["task"]).pop()
        mi = mi.append(bestmods.mean().append(pd.Series({"model_name": mname + "glomean",
                                                         "nodes_per_grid": npg, "task": task})), ignore_index=True)
        mi = mi.append(bestmods.sem().append(pd.Series({"model_name": mname + "glosem", 
                                                        "nodes_per_grid": npg, "task": task})), ignore_index=True)
    

# For a more fine grained analysis look at mean per data size
for mname in mnames:
    if mname + "mean" in set(mi["model_name"]): 
        break
    for idx in indices:
        mods = mi[(mi["model_name"] == mname) & (mi["train_idx"] == idx)]
        # Assert that all have the same number of nodes
        assert len(list(set(mods["nodes_per_grid"]))) == 1
        npg = set(mods["nodes_per_grid"]).pop()
        task = set(mods["task"]).pop()
        bestmods = mods[metrics].sort_values("ev"+str(int(npg))+"_r2").tail(3)
        #mi = mi.drop(mods.index)
        mi = mi.append(bestmods.mean().append(pd.Series({"train_idx": idx, 
                                                         "model_name": mname + "mean",
                                                         "nodes_per_grid": npg, "task": task})), ignore_index=True)
        mi = mi.append(bestmods.sem().append(pd.Series({"train_idx": idx, 
                                                        "model_name": mname + "sem", 
                                                        "nodes_per_grid": npg, "task": task})), ignore_index=True)
    
```

```python
"""
Find ints in a string, check to which dataset the number of nodes corresponds, return number of grids
"""
def intmatch(x):
    a = re.search(r"(?<=stop=)\d+", str(x))
    if a:
        b = int(a.group())
        if b in [700, 7000, 70000, 700000]:
            return b/100
        else:
            return b/ 20
    else:
        return np.nan

mi = mi.astype({"nodes_per_grid": float})

# Assert that this returns the same thing!!!
#assert 
#mi["model_name"][((mi["nodes_per_grid"] == np.array(list(map(intmatch, mi["train_idx"])))).to_numpy() ==  False)]


mi["num_grids_for_training"] = np.array(list(map(intmatch, mi["train_idx"])))
```

```python
def get_sparse_predictors(submi):
    L = []
    for l in list(map(lambda x: x.split("'"), submi["sparse_predictors"].replace(np.nan,""))):
        for ll in l:
            L.append(ll)

    L = [l for l in L if not l.startswith(",")]
    L = [l for l in L if not l.startswith("Ind")]
    L = [l for l in L if not l.startswith("object")]
    L = [l for l in L if not l.startswith("]")]
    L = [l for l in L if not l.startswith(")")]
    return L
```

```python
submi20 = mi[(mi["task"] == "SNBS") & (mi["nodes_per_grid"] == 20)]
submi20 = submi20[list(map(lambda x: x.startswith("omp"), submi20["model_name"]))]

submi100 = mi[(mi["task"] == "SNBS") & (mi["nodes_per_grid"] == 100)]
submi100 = submi100[list(map(lambda x: x.startswith("omp"), submi100["model_name"]))]

l20 = get_sparse_predictors(submi20)
l100 = get_sparse_predictors(submi100)
```

```python
def add_model_to_axes(ax, model_name, label, x=[7,70,700,7000], sub_index20=[0,1,2,3], sub_index100=[0,1,2,3], verbose_legend=False, **kwargs):
    y20 = mi[(mi["model_name"] == model_name) & 
        (mi["nodes_per_grid"] == 20)].sort_values(
        by=["num_grids_for_training"])["ev20_r2"].to_numpy().astype(float)
    y100 = mi[(mi["model_name"] == model_name) & 
        (mi["nodes_per_grid"] == 100)].sort_values(
        by=["num_grids_for_training"])["ev100_r2"].to_numpy().astype(float)
    if verbose_legend:
        ax.plot(np.array(x)[sub_index20]*20, y20[sub_index20], label=label+"20", linestyle = "dashed", **kwargs)
        ax.plot(np.array(x)[sub_index100]*100, y100[sub_index100], label=label+"100", linestyle = "solid", **kwargs)
    else:
        ax.plot(np.array(x)[sub_index20]*20, y20[sub_index20], label=None, linestyle = "dashed", **kwargs)
        ax.plot(np.array(x)[sub_index100]*100, y100[sub_index100], label=label, linestyle = "solid", **kwargs)        
    
    return ax
```

```python
def intmatch(x):
    a = re.search(r"(?<=stop=)\d+", str(x))
    if a:
        b = int(a.group())
        if b in [700, 7000, 70000, 700000]:
            return b/100
        else:
            return b/ 20
    else:
        return np.nan

mi = mi.astype({"nodes_per_grid": float})

# Assert that this returns the same thing!!!
#assert 
#mi["model_name"][((mi["nodes_per_grid"] == np.array(list(map(intmatch, mi["train_idx"])))).to_numpy() ==  False)]


mi["num_grids_for_training"] = np.array(list(map(intmatch, mi["train_idx"])))
```

```python
def plot_performance(model, metric, save=False, savepath=None, title=True):
    x = np.array([7,70,700,7000])
    y_arr =[
        mi[(mi["model_name"] == model) & 
        (mi["nodes_per_grid"] == 20)].sort_values(
        by=["num_grids_for_training"])["ev20_" + metric].to_numpy().astype(float),
        mi[(mi["model_name"] == model) & 
        (mi["nodes_per_grid"] == 20)].sort_values(
        by=["num_grids_for_training"])["ev100_" + metric].to_numpy().astype(float),
        mi[(mi["model_name"] == model) & 
        (mi["nodes_per_grid"] == 20)].sort_values(
        by=["num_grids_for_training"])["evTex_" + metric].to_numpy().astype(float),
        mi[(mi["model_name"] == model) & 
        (mi["nodes_per_grid"] == 100)].sort_values(
        by=["num_grids_for_training"])["ev20_" + metric].to_numpy().astype(float),
        mi[(mi["model_name"] == model) & 
        (mi["nodes_per_grid"] == 100)].sort_values(
        by=["num_grids_for_training"])["ev100_" + metric].to_numpy().astype(float),
        mi[(mi["model_name"] == model) & 
        (mi["nodes_per_grid"] == 100)].sort_values(
        by=["num_grids_for_training"])["evTex_" + metric].to_numpy().astype(float),]

    plt.figure(dpi=300)
    plt.plot(x * 20,y_arr[0], label="tr20ev20", linestyle = "dotted", marker = "^", color="tab:blue")
    plt.plot(x * 100,y_arr[4], label="tr100ev100", linestyle = "solid", marker = "^", color="tab:blue")
    plt.plot(x * 20,y_arr[1], label="tr20ev100", linestyle = "dotted", marker = "v", color = "tab:cyan")
    plt.plot(x * 20,y_arr[2], label="tr20evTex", linestyle = "dotted", marker = "o", color="tab:pink")
    #plt.plot(x,y_arr[3], label="tr100ev20", linestyle = "dashdot", marker = "x")

    plt.plot(x * 100,y_arr[5], label="tr100evTex", linestyle = "solid", marker = "o", color="tab:pink")
    plt.xscale("log")
    plt.xlabel("nodes for training")
    plt.ylabel(metric.upper())
    plt.ylim([0,1])
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    if title:
        plt.title(metric.upper() + " for " + model)
    if save == True:
        plt.savefig(model + "_" + metric.upper() + ".png")
    if savepath:
        plt.savefig(savepath)
    plt.show()
```
